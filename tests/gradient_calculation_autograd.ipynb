```python
import torch
```

```python
x = torch.randn(3, requires_grad=True)
print(x)
```

```python
y = x + 2
print(y)
```

```python
z = y*y*2
z = z.mean()
print(z)
```

```python
z.backward()  # dz/dx
print(x.grad)
```

```python
weights = torch.ones(4, requires_grad=True)
```
