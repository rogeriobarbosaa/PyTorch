```python
import torch
```

```python
# dataset: feature e label
x = torch.tensor(1.0)
y = torch.tensor(2.0)
```

```python
# peso
w = torch.tensor(1.0, requires_grad=True)
```

```python
# previsão
y_pred = w * x
```

```python
# loss (mse)
loss = (y_pred - y)**2
print(loss)
```

```python
# backward pass
loss.backward()  # de/dw = de/dy_pred * dy_pred/dw
print(w.grad)
```
