```python
# gpu
if torch.cuda.is_available():         # se gpu está disponível
    device = torch.device("cuda")     # salva o serviço: gpu
    x = torch.ones(5, device=device)  # define o serviço onde será carregado e rodará: gpu
    y = torch.ones(5)
    y = y.to(device)                  # define o serviço onde será carregado e rodará: gpu
    z = x + y
    z = z.to("cpu")                   # define o serviço onde será carregado e rodará: cpu
```
